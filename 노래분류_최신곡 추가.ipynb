# 1. 가사 전처리

```python
import pandas as pd

df = pd.read_csv('all_plus_clean.csv')
df
```

```python
from hanspell import spell_checker
from pykospacing import spacing
from soynlp.normalizer import *
from konlpy.tag import Okt

okt = Okt()
```

```python
# 노래 감정 사전
emotion = []

f = open('감정사전_사랑좋다제거_명사제거_긍정단어추가.csv', encoding='UTF-8')
lines = f.readlines()

for line in lines:
    line = line.strip()
    emotion.append(line)
f.close()
```

```python
from tqdm.notebook import tqdm

# 전처리 함수
# 노래가사는 띄어쓰기, 맞춤법 검사, 중복 수 제한 단계 필요 없음
def text_preprocessing(textlist):
    X_train = []
    i=0
    for sentence in tqdm(textlist):
        try:
            temp_X = []
            new_sentence = spacing(sentence)
            temp_X = okt.morphs(new_sentence, norm=True, stem=True)
            temp_X = [w for w in temp_X if w in stopwords]
            X_train.append(temp_X)
        except Exception as e:
            temp_X = ['에러']
            X_train.append(temp_X)
            print(i,'번째에서 error 발생', e)
        i+=1
    return X_train
```

```python
df['lyric_clean'] = text_preprocessing(df['lyric'])
df
```

```python
# id, 장르, 제목, 가수, 가사, 가사 전처리한 가사 순서로 재배치

df1 = df.reindex(columns=['_id', 'genre', 'title', 'singer', 'lyric', 'lyric_clean'])
df1
```

```python
df1.to_csv('노래추가_전처리.csv')
```

# 2. 노래 분류

```python
from gensim import corpora

dictionary = corpora.Dictionary(df1['lyric_clean'])
corpus = [dictionary.doc2bow(text) for text in df1['lyric_clean']]
print(corpus[1])    #수행된 결과에서 두번째 가사 출력 (첫번째 가사의 인덱스는 0)
```

```python
import gensim

NUM_TOPICS = 2    #두개의 토픽으로 분류
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
topics = ldamodel.print_topics(num_words=4)    #주제와 연관성 높은 단어 4개 뽑기

for topic in topics:
    print(topic)
```

```python
import pyLDAvis.gensim

#시각화 단계
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
pyLDAvis.display(vis)
```

```python
import pandas as pd

def make_topictable_per_doc(ldamodel, corpus):
    topic_table = pd.DataFrame()

    # 몇 번째 문서인지를 의미하는 문서 번호와 해당 문서의 토픽 비중을 한 줄씩 꺼내온다.
    for i, topic_list in enumerate(ldamodel[corpus]):
        doc = topic_list[0] if ldamodel.per_word_topics else topic_list            
        doc = sorted(doc, key=lambda x: (x[1]), reverse=True)
        # 각 노래가사에 대해서 비중이 높은 토픽순으로 토픽을 정렬한다.
        # EX) 정렬 전 0번 노래가사 : (2번 토픽, 48.5%), (8번 토픽, 25%), (10번 토픽, 5%), (12번 토픽, 21.5%), 
        # Ex) 정렬 후 0번 노래가사 : (2번 토픽, 48.5%), (8번 토픽, 25%), (12번 토픽, 21.5%), (10번 토픽, 5%)
        # 48.5% > 25% > 21.5% > 5% 순으로 정렬이 된 것.

        # 모든 노래가사에 대해서 각각 아래를 수행
        for j, (topic_num, prop_topic) in enumerate(doc): #  몇 번 토픽인지와 비중을 나눠서 저장한다.
            if j == 0:  # 정렬을 한 상태이므로 가장 앞에 있는 것이 가장 비중이 높은 토픽
                topic_table = topic_table.append(pd.Series([int(topic_num), round(prop_topic,4), topic_list]), ignore_index=True)
                # 가장 비중이 높은 토픽과, 가장 비중이 높은 토픽의 비중과, 전체 토픽의 비중을 저장한다.
            else:
                break
    return(topic_table)
```

```python
topictable = make_topictable_per_doc(ldamodel, corpus)
topictable.columns = ['가장 비중이 높은 토픽', '가장 높은 토픽의 비중', '각 토픽의 비중']
topictable[:30]
```

```python
#라벨링화
label = topictable['가장 비중이 높은 토픽']
label
```
