```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import pandas as pd
import time as time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
# driver.get("https://www.melon.com/genre/song_list.htm?gnrCode=GN0100&steadyYn=Y")
```

```python
def func(driver, target):
    
    page = [] # 음원 page가 담길 list
    LYRIC = [] # 가사가 담길 list
#   page의 주소가 1, 51, 101로 구분되어서 만들어주었다.
    page_ = target // 50
    page.append(1)
    for i in range(1, page_+1):
        page_num = 1 + 50 * i
        page.append(page_num)
#     print(page)
    
    count = 0 # 종료 시킬 카운트
    
    j = 0 # page list의 index
    while target >= len(LYRIC):
        #   1.페이지 이동
        page_url = "https://www.melon.com/genre/song_list.htm?gnrCode=GN0100&steadyYn=Y#params%5BgnrCode%5D=GN0100&params%5BdtlGnrCode%5D=&params%5BorderBy%5D=NEW&params%5BsteadyYn%5D=Y&po=pageObj&startIndex={}".format(page[j])
        driver.get(page_url)
        j += 1 
        time.sleep(3)
        
    #    2. 각 음원 상세보기 주소 긁어 오기
        title = driver.find_elements_by_class_name("input_check")
        # title =  driver.find_elements_by_class_name('div.wrap t_right')

        title2 = []

        for i in title:
            title2.append(i.get_attribute("value"))
#         print(title2)
        time.sleep(3)
        
        # 3. 2에서 가져온 주소를 활용해 가사 크롤링
        for i in title2[1:]:
            driver.get("https://www.melon.com/song/detail.htm?songId=" + i)
            lyric = driver.find_element_by_class_name("lyric")
            LYRIC.append(lyric.text)
            count += 1
            
        if count == target:
            break
            
    return LYRIC
```

```python
song_list = func(driver, 100)
song_list
```

```python
df = pd.DataFrame()
df['가사'] = song_list
df
```
